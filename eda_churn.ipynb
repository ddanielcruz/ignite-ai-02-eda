## Bivariate analysis

### Hipótese 1 - Clientes com contrato mensal têm maior probabilidade de cancelamento

```python
# Cross-tabulate "Churn" and "Contract" columns (tabela de contingência)
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total')
```

```python
pd.crosstab(df_churn.Churn, df_churn.Contract, normalize="index", margins=True, margins_name='Total')
```

Considerando o público total, os clientes com contrato mensal representam 55%. Porém, se considerarmos apenas os clientes que abandonaram o serviço, 88% deles tinham contrato mensal.

#### Correlação de variáveis qualitativas

Para avaliar a correlação entre duas variáveis categóricas (qualitativas), devemos executar um teste de hipótese chamado Chi-Square ou Qui-Quadrado de Pearson. Em um teste de hipóteses, duas hipóteses são formuladas:

- H0 (hipótese nula): as duas variáveis são independentes
- H1 (hipótese complementar): as duas variáveis são complementares

O teste serve para confirmar ou recusar a hipótese nula (H0). Quando a probabilidade de observarmos H0 é inferior a 0.05 (p=value), recusamos a hipótese nula e seguimos com a complementar.

```python
df_crosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
# Calculate scores and p-values
from scipy.stats import chi2_contingency

chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
p_values_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
pd.set_option('display.float_format', '{:.15f}'.format)
```

```python
df_chi_scores_churn_contract = pd.DataFrame({'chi2': scores_churn_contract, 'p-value': p_values_churn_contract})
df_chi_scores_churn_contract
```

Conforme o teste estatístico (Qui-Square), o p-value é menor ou igual a 0.05. Pelo qui2 alto, podemos afirmar que há uma forte correlação. Portanto, seguimos com a hipótese complementar.

### Tukey Method

The Tukey method is based on the interquartile range (IQR), which is the difference between the third quartile (Q3) and the first quartile (Q1). The IQR is used to define the lower and upper limits of the data distribution. Any value outside these limits is considered an outlier. It's used for **non-normal distributions.**

```python
# Método de Tukey para detecção de outliers. O método de Tukey é baseado no intervalo interquartil (IQR), aplicado em distribuições não-normais.
# IQR = Q3 - Q1
Q1_TotalCharges = df_churn_monthly.TotalCharges.quantile(0.25)
Q3_TotalCharges = df_churn_monthly.TotalCharges.quantile(0.75)
IQR_TotalCharges = Q3_TotalCharges - Q1_TotalCharges
IQR_TotalCharges
```

```python
# Limite inferior = Q1 - 1.5 * IQR
inferior_limit_TotalCharges = Q1_TotalCharges - 1.5 * IQR_TotalCharges

# Limite superior = Q3 + 1.5 * IQR
superior_limit_TotalCharges = Q3_TotalCharges + 1.5 * IQR_TotalCharges

inferior_limit_TotalCharges, superior_limit_TotalCharges
```

```python
df_churn_monthly[(df_churn_monthly.TotalCharges < inferior_limit_TotalCharges) | (df_churn_monthly.TotalCharges > superior_limit_TotalCharges)]
```

### Z-Score

The Z-Score is a statistical measurement that describes a value's relationship to the mean of a group of values. It is measured in terms of standard deviations from the mean. If the Z-Score is greater than 3 or less than -3, the data point is considered an outlier. It's used for **normally distributed data.**

```python
from scipy.stats import zscore
```

```python
z_scores_TotalCharges = np.abs(zscore(df_churn_monthly.TotalCharges))
z_scores_TotalCharges
```

```python
df_churn_monthly[z_scores_TotalCharges > 3.0]
```
